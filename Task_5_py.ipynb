## Load and clean data

### Subtask:
Load the dataset and perform necessary data cleaning and preparation steps.


**Reasoning**:
Load the dataset, display the first 5 rows, check for missing values, and display data types to understand the data structure and identify potential cleaning needs.



```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/superstore.csv")

# Display the first 5 rows
print("First 5 rows of the DataFrame:")
display(df.head())

# Check for missing values
print("\nMissing values per column:")
display(df.isnull().sum())

# Display data types
print("\nData types of each column:")
display(df.dtypes)
```

**Reasoning**:
The previous step loaded the data and checked for missing values and data types. There are no missing values. The 'Order.Date' and 'Ship.Date' columns are of object type and should be converted to datetime objects for potential time-based analysis.



```python
# Convert 'Order.Date' and 'Ship.Date' to datetime objects
df['Order.Date'] = pd.to_datetime(df['Order.Date'])
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])

# Display the data types again to confirm the conversion
print("Data types after converting date columns:")
display(df.dtypes)
```

## Build streamlit app structure

### Subtask:
Create the basic structure for the Streamlit application, including a title and a sidebar for filters.


**Reasoning**:
Create the basic structure of the Streamlit app, including the title and sidebar header.



```python
import streamlit as st

st.title("Superstore Sales Dashboard")
st.sidebar.header("Filters")
```

## Implement filters

### Subtask:
Add interactive filters for Region, Category, and Sub-Category to the sidebar.


**Reasoning**:
Add interactive filters for Region, Category, and Sub-Category to the sidebar using st.sidebar.multiselect and then filter the dataframe based on the selected values.



```python
import streamlit as st

# Create multiselect filters for Region, Category, and Sub-Category
selected_regions = st.sidebar.multiselect(
    "Select Region",
    df['Region'].unique().tolist()
)
selected_categories = st.sidebar.multiselect(
    "Select Category",
    df['Category'].unique().tolist()
)
selected_subcategories = st.sidebar.multiselect(
    "Select Sub-Category",
    df['Sub.Category'].unique().tolist()
)

# Filter the DataFrame based on selections
filtered_df = df.copy()

if selected_regions:
    filtered_df = filtered_df[filtered_df['Region'].isin(selected_regions)]
if selected_categories:
    filtered_df = filtered_df[filtered_df['Category'].isin(selected_categories)]
if selected_subcategories:
    filtered_df = filtered_df[filtered_df['Sub.Category'].isin(selected_subcategories)]

# Display the head of the filtered DataFrame to verify
display(filtered_df.head())
```

## Calculate and display kpis

### Subtask:
Calculate and display key performance indicators (KPIs) such as Total Sales and Profit.


**Reasoning**:
Calculate the total sales and profit from the filtered data and display them as KPIs using Streamlit's st.metric.



```python
import streamlit as st

# Calculate total sales and profit
total_sales = filtered_df['Sales'].sum()
total_profit = filtered_df['Profit'].sum()

# Display KPIs
st.metric("Total Sales", f"${total_sales:,.2f}")
st.metric("Total Profit", f"${total_profit:,.2f}")
```

## Visualize top 5 customers by sales

### Subtask:
Create a chart to visualize the top 5 customers by sales.


**Reasoning**:
Create a chart to visualize the top 5 customers by sales by grouping the filtered data by customer, summing their sales, sorting, selecting the top 5, and creating an Altair bar chart.



```python
import altair as alt
import streamlit as st

# Group by customer and sum sales
customer_sales = filtered_df.groupby('Customer.Name')['Sales'].sum().reset_index()

# Sort by sales and get top 5
top_5_customers = customer_sales.sort_values(by='Sales', ascending=False).head(5)

# Create the bar chart
chart = alt.Chart(top_5_customers).mark_bar().encode(
    x=alt.X('Customer.Name', sort='-y'),
    y='Sales',
    tooltip=['Customer.Name', 'Sales']
).properties(
    title='Top 5 Customers by Sales'
)

# Display the chart in Streamlit
st.altair_chart(chart, use_container_width=True)
```

*   Consider adding more visualizations, such as sales trends over time, profit by category, or a map visualization of sales by region, to provide a more comprehensive view of the business performance.


```python
%%writefile app.py
import pandas as pd
import streamlit as st
import altair as alt

# Load the dataset
df = pd.read_csv("/content/superstore.csv")

# Convert 'Order.Date' and 'Ship.Date' to datetime objects
df['Order.Date'] = pd.to_datetime(df['Order.Date'])
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])

st.title("Superstore Sales Dashboard")
st.sidebar.header("Filters")

# Create multiselect filters for Region, Category, and Sub-Category
selected_regions = st.sidebar.multiselect(
    "Select Region",
    df['Region'].unique().tolist()
)
selected_categories = st.sidebar.multiselect(
    "Select Category",
    df['Category'].unique().tolist()
)
selected_subcategories = st.sidebar.multiselect(
    "Select Sub-Category",
    df['Sub.Category'].unique().tolist()
)

# Filter the DataFrame based on selections
filtered_df = df.copy()

if selected_regions:
    filtered_df = filtered_df[filtered_df['Region'].isin(selected_regions)]
if selected_categories:
    filtered_df = filtered_df[filtered_df['Category'].isin(selected_categories)]
if selected_subcategories:
    filtered_df = filtered_df[filtered_df['Sub.Category'].isin(selected_subcategories)]

# Calculate total sales and profit
total_sales = filtered_df['Sales'].sum()
total_profit = filtered_df['Profit'].sum()

# Display KPIs
st.metric("Total Sales", f"${total_sales:,.2f}")
st.metric("Total Profit", f"${total_profit:,.2f}")

# Group by customer and sum sales
customer_sales = filtered_df.groupby('Customer.Name')['Sales'].sum().reset_index()

# Sort by sales and get top 5
top_5_customers = customer_sales.sort_values(by='Sales', ascending=False).head(5)

# Create the bar chart
chart = alt.Chart(top_5_customers).mark_bar().encode(
    x=alt.X('Customer.Name', sort='-y'),
    y='Sales',
    tooltip=['Customer.Name', 'Sales']
).properties(
    title='Top 5 Customers by Sales'
)

# Display the chart in Streamlit
st.altair_chart(chart, use_container_width=True)
```
